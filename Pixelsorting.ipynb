### Fun way to generate simple unique avatars/icons.

```python
import numpy as np
import os
import matplotlib.pyplot as plt
from skimage.transform import resize
import utils

Rpath = "images"
Wpath = "pixels"

def get_images(path):
    files = [os.path.join(path, fname)
        for fname in os.listdir(path)]
    imgs = [plt.imread(fname)[..., :3] for fname in files]
    imgs = [resize(img_i, (32, 32)) for img_i in imgs]
    imgs = np.array(imgs).astype(np.float32)
    imgs = [np.sort(img_i, axis=0) for img_i in imgs]
    imgs = [resize(img_i, (16, 16)) for img_i in imgs]
    return np.array(imgs)
```

```python
imgs = get_images(Rpath)
```

```python
def save_icons():
    for i in range(0, len(imgs)):
    
        arr = imgs[i]
        f = 'pixels/000%02d.jpg' % i
        plt.imsave(fname=f, arr=arr)
```

```python
save_icons()
```
